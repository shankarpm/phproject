####1a. Total number of flights by Airlines on a monthly basis

```
sdf_airlines= spark.read \
  .format("snowflake") \
  .options(**options) \
  .option("query",  "select * from USER_SHANKAR.PUBLIC.V_TOTAL_FLIGHTS_AIRLINE") \
  .load()

airline_month_df=sdf_airlines.toPandas()
airline_month_df['MONTH'] = airline_month_df['MONTH'].astype(int).apply(lambda x: calendar.month_abbr[x])
display(airline_month_df)
```

### 5. Delay reason by airport

```
sdf_delay_reason= spark.read \
  .format("snowflake") \
  .options(**options) \
  .option("query",  "select * from USER_SHANKAR.PUBLIC.V_DELAY_REASON_AIRPORT") \
  .load()
delay_reason_df=sdf_delay_reason.toPandas()
delay_reason_df = delay_reason_df.head(20)
display(delay_reason_df)
```

```
fig, ax1 = plt.subplots(figsize=(30, 10))
#sns.set(rc={'figure.figsize':(20,10)}) 
tidy = delay_reason_df.melt(id_vars='AIRPORT').rename(columns=str.title)
bp = sns.barplot(x='Airport', y='Value', hue='Variable', data=tidy, ax = ax1)
bp.set_title("Delay reason by airport")
bp.set(ylabel='Delay Count')
for item in bp.get_xticklabels():
    item.set_rotation(80)
```
